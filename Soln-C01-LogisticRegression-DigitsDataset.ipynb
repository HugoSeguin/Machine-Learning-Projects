We begin by importing the packages and modules that we will need.

```python
import numpy as np 
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from common_plots import DigitPlotter
```

**1. Load the digits dataset from scikit-learn**

```python
digits = load_digits()
print(digits.keys())
print("data shape: " + str(np.shape(digits.data)))
print("target shape: " + str(np.shape(digits.target)))
print("feature names: " + str(digits.feature_names))
print("target_names: " + str(digits.target_names))
```

Plot the data

```python
num_plots = 6
plotter = DigitPlotter((8,8))
plotter.is_binary_cm = False
plotter.plot_multiple(digits.data[0:num_plots], digits.target[0:num_plots])
```

**2. Split the data into training and test sets such that 20% of the dataset will be used as the test set**

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=0)
```

**3. Use LogisticRegression to train the model**

```python
clf = LogisticRegression(max_iter=10000).fit(x_train, y_train)
```

**4. Make predictions on the test set**

```python
y_pred = clf.predict(x_test)
```

**5. Compute the fraction of correctly classified samples in the test set**

```python
print(metrics.accuracy_score(y_test, y_pred))
print(clf.score(x_test, y_test))
```
